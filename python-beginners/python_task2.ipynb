Дано: 
* функция parse_words(text), которая разбирает текст (text) на слова (возвращает список).
* текст (юникод-строка)
* список русских союзов 

```python
import re
def parse_words(text):
    return re.findall(r'\w+', text, re.UNICODE)

txt = u'''
Наконец я возвратился из моей двухнедельной отлучки. Наши уже три дня как были в Рулетенбурге. 
Я думал, что они и бог знает как ждут меня, однако ж ошибся. Генерал смотрел чрезвычайно независимо, поговорил со мной свысока и отослал меня к сестре. 
Было ясно, что они где-нибудь перехватили денег. Мне показалось даже, что генералу несколько совестно глядеть на меня. 
Марья Филипповна была в чрезвычайных хлопотах и поговорила со мною слегка; деньги, однако ж, приняла, сосчитала и выслушала весь мой рапорт. 
К обеду ждали Мезенцова, французика и еще какого-то англичанина: как водится, деньги есть, так тотчас и званый обед, по-московски. 
Полина Александровна, увидев меня, спросила, что я так долго? и, не дождавшись ответа, ушла куда-то. Разумеется, она сделала это нарочно. 
Нам, однако ж, надо объясниться. Много накопилось. 
'''

conjunctions = u"""абы аж ажно ай аки ако али аль ан аще ж же зане зато зачем значит и или 
иль именно имже иначе инда итак кабы как как-то каков какой ковда ковды когда когды коли 
коль который куда ли либо лишь однако почему почём правда прежде притом причём просто пускай 
пусть раз раз разве разве ровно ровно сиречь сколько следовательно словно также теперь то хоть 
хотя чем чем чи что чтоб чтобы чуть яко якобы
"""

conjunctions = conjunctions.split()
```

Задание:
* составить словарь, где ключ - слово из текста, значение - количество употреблений в тексте
* составить список слов из текста без дубликатов
* найти союзы из списка, встречающихся в тексте, и количество их употреблений 

### Решение

Первым делом применим функцию parse_words к тексту, чтобы получить список слов:

```python
allwords = parse_words(txt)
```

```python
for w in allwords: print w
```

Имея теперь список слов из текста, сформируем словарь со статистикой употребляемости:

```python
d = {}
for w in allwords:
    if d.has_key(w):
        d[w] += 1
    else:
        d[w] = 1
```

Проверим результат, напечатав все слова, встречающиеся больше раза:

```python
for w in d:
    if d[w] > 1:
        print d[w], w
```

Теперь нужно получить список слов без повторений. Легко заметить, что теперь ключи словаря со статистикой и будут тем списком:

```python
unique_words = d.keys()
```

```python
for w in sorted(unique_words): print w, 
```

Но если бы мы не имели готовый словарь, то самым простым способом получить список уникальных слов - преобразовать к set и обратно к list:

```python
unique_words_new = list(set(allwords))
```

```python
unique_words == unique_words_new
```

Теперь нам необходимо собрать статистику использования союзов в тексте, имея список самих союзов:

```python
for w in sorted(conjunctions): print w,
```

Конечно можно еще раз пройтись по всем словам, добавив дополнительное условие, проверяющее слово на пренадлежность к союзам, но мы снова воспользуемся типом set:
- преобразуем список союзов к типу set
- найдем пересечение с уникальными словами из текста
- составим аналогичный словарь, на основе уже собранной статистики

```python
conjunctions = set(conjunctions)
```

```python
d_new = {}
for w in conjunctions.intersection(unique_words):
    d_new[w] = d[w]
```

```python
for k in d_new:
    print d_new[k], k
```
